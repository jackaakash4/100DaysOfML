```python
#Feature engineering
#Log transaction amount
import numpy as np
df['LogAmount'] = np.log1p(df['Amount'])
df.head()
```

```python
#visualize the log amount
sns.histplot(
    data = df,
    x = 'LogAmount'
)
plt.show()
```

```python
#visualizing amount without log to see difference
sns.histplot(
    data = df,
    x = 'Amount'
)
plt.show()
#loging reduces the skewness
```

```python
#Feature engineering: Amount Bins
df['AmountBin'] = pd.qcut(
    df['Amount'],
    q = 5,
    labels = False
)
df.head()
```

```python
sns.countplot(
    data = df,
    x = 'AmountBin'
)
plt.show()
#Equally distributed into four categories
```

```python
#Feature engineering: High value transaction

threshold = (
    df['Amount']
    .quantile(0.95)
)

df['HighValue'] = (
    df['Amount'] > threshold
).astype(int)
df.head()
```
