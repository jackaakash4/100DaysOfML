```python
import numpy as np
import matplotlib.pyplot as plt

#define function
def f(x):
    return x**2

def grad(x):
    return 2*x

#initialize
x = 20
learning_rate = 0.1
iterations = 30

#for visualization
history = []
for i in range(iterations):
    gradient = grad(x)
    history.append(x)
    x = x - learning_rate * gradient
    print(f"{i+1}\t{x}\t{f(x)}")
```

```python
history
```

```python
#for visualization
xs = np.linspace(-1, 30, 100)
ys = xs ** 2

plt.plot(xs, ys, label = "f(x)=x^2")
plt.scatter(history,[h**2 for h in history], color = "red", label = "Gradient Descent Step")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradiend Descent toward the minimun")
plt.legend()
plt.show()
```
